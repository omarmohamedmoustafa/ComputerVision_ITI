```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    stat , frame =cap.read()
    if stat ==True:
        cv2.imshow("my_video",frame)
        k =cv2.waitKey(16)
        if k == ord('q'):
            break
    else:
        print("wrong frame")
cap.release()
cv2.destroyAllWindows()            
```
