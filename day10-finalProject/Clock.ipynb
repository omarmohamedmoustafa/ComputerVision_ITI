```python
import cv2
import numpy as np
from datetime import date , datetime
from math import sin , cos , pi
```

```python
datenw = date.today()
radius = 400
centerX=400
centerY=400
img = np.zeros((800,800,3),np.uint8)
cv2.circle (img,(400,400),300,(0,0,255),2)
for i in range (60):
    sec= i*6-90
    x1=300*cos(sec*pi/180)+radius
    y1=300*sin(sec*pi/180)+radius
    x1= np.uint16(x1)
    y1= np.uint16(y1)
    cv2.circle(img,(x1,y1),3,(255,255,255),-1)
    x=str(i/5)
    y=str(x[0])
    z=str(x[1])
    if i==0:
        cv2.putText(img,'12',(x1-15,y1-10),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
    elif (i%5==0) and (i>=50):
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,y+z,(x1-25,y1-5),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif(i%5==0) and (0<i<=10):
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,y,(x1+5,y1-5),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif(i%5==0) and (15<i<30):
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,y,(x1+5,y1+20),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif i ==30:
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,'6',(x1-5,y1+20),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif i ==15:
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,'3',(x1+5,y1+5),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif i ==45:
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,'9',(x1-20,y1+5),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    elif(i%5==0) and (30<i<45):
        cv2.circle(img,(x1,y1),4,(255,255,255),-1)
        cv2.putText(img,y,(x1-20,y1+20),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)

while True:
    cv2.circle(img,(400,400),4,(255,255,255),-1)
    
    img[:60,:250]=[0,0,0]
    
    cv2.putText(img,'Date: '+ str(datenw),(20,20),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
    cv2.putText(img,'Time: '+ str(datetime.now().strftime("%H:%M:%S")),(20,40),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,255,0),1)
 
    x=float (datetime.now().strftime("%S"))
    y=float(datetime.now().microsecond)/1000000    
    sec=x+y
    sec = sec*6-90
    x1=280*cos(sec*pi/180)+radius
    y1=280*sin(sec*pi/180)+radius
    x1= np.uint32(x1)
    y1= np.uint32(y1)
    min= ((int(datetime.now().strftime("%M"))+int(datetime.now().strftime("%S"))/60))*6-90
    x2=250*cos(min*pi/180)+radius
    y2=250*sin(min*pi/180)+radius
    x2= np.uint32(x2)
    y2= np.uint32(y2)  

    hrs= ((int(datetime.now().strftime("%H"))+int(datetime.now().strftime("%M"))/60))*6*5-90
    x3=200*cos(hrs*pi/180)+radius
    y3=200*sin(hrs*pi/180)+radius
    x3= np.uint32(x3)
    y3= np.uint32(y3)

    cv2.line(img,(400,400),(x1 ,y1),(0,0,255),1)
    cv2.line(img,(400,400),(x2 ,y2),(0,255,0),2)
    cv2.line(img,(400,400),(x3 ,y3),(255,0,0),3)
    cv2.putText(img,'Omartz',(360,600),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1,(0,0,100),1)
    cv2.circle(img,(460,590),8,(255,255,255))
    cv2.putText(img,'R',(455,595),cv2.FONT_HERSHEY_PLAIN ,1,(0,0,255),1)
    cv2.imshow('My_Clock',img)
    cv2.line(img,(400,400),(x1,y1),(0,0,0),1)
    cv2.line(img,(400,400),(x2 ,y2),(0,0,0),2)
    cv2.line(img,(400,400),(x3 ,y3),(0,0,0),3)
    k=cv2.waitKey(1)
    
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
