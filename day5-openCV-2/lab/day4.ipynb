```python
import cv2
cap=cv2.VideoCapture("challenge.mp4")

while cap.isOpened():
    stat, frame=cap.read()
    if stat==True:
        cv2.imshow("my_video",frame)
        k=cv2.waitKey(10)
        print(k)
        if k==ord('q'):
            break
    else:
        cv2.destroyAllWindows()
        
        
        print("wrong_frame")
        break
cap.release()
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np


def nothing(x):
    print(x)
cv2.namedWindow("my_image")
cv2.createTrackbar('R','my_image',0,255,nothing)
cv2.createTrackbar('G','my_image',0,255,nothing)
img=np.zeros((500,500,1),np.uint8)
while True:
    r=cv2.getTrackbarPos('R','my_image')
    img[:]=r
    cv2.imshow("my_image",img)# BGR
    k=cv2.waitKey(1000)
    if k==ord('q'):
        break
cv2.destroyAllWindows()

```
